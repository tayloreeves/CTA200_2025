```python
import numpy as np
import matplotlib.pyplot as plt
from ass3 import ass3_func

# Initial Parameters
xmin, xmax, ymin, ymax = -2, 2, -2, 2
width, height = 1000, 1000
max_iter = 100

# Generate data set
'''
This function takes in the data provided about the initial parameters, and pushes them through the function within the module.
'''
data = ass3_func(xmin, xmax, ymin, ymax, width, height, max_iter)

# Plot 1: The black portions of the graph are where the set is divergent, and the white portions is where is is bounded
plt.figure(figsize=(8, 8))
plt.imshow(data == 0, extent=(xmin, xmax, ymin, ymax), cmap='binary')
plt.title("Mandelbrot Set (White=Bounded, Black=Divergent")
plt.xlabel("Real Portion of term")
plt.ylabel("Imaginary Portion of term")
plt.show()

# Plot 2: This map shows where it diverges off to infinity, and how many iterations happened in.
plt.figure(figsize=(8, 8))
plt.imshow(data, extent=(xmin, xmax, ymin, ymax), cmap='turbo')
plt.title("Mandelbrot Set, with colours to showwhen values begin to bound")
plt.xlabel("Real Portion of term")
plt.ylabel("Imaginary Portion of term")
plt.colorbar(label='Iteration at Divergence, as a decimal of the total iterations (100)')
plt.show()
```
